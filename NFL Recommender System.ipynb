```python
nfl_teams = ['Arizona Cardinals',
'Atlanta Falcons',
'Baltimore Ravens',
'Buffalo Bills',
'Carolina Panthers',
'Chicago Bears',
'Cincinnati Bengals',
'Cleveland Browns',
'Dallas Cowboys',
'Denver Broncos',
'Detroit Lions',
'Green Bay Packers',
'Houston Texans',
'Indianapolis Colts',
'Jacksonville Jaguars',
'Kansas City Chiefs',
'Las Vegas Raiders',
'Los Angeles Chargers',
'Los Angeles Rams',
'Miami Dolphins',
'Minnesota Vikings',
'New England Patriots',
'New Orleans Saints',
'New York Giants',
'New York Jets',
'Philadelphia Eagles',
'Pittsburgh Steelers',
'San Francisco 49ers',
'Seattle Seahawks',
'Tampa Bay Buccaneers',
'Tennessee Titans',
'Washington Football Team']
nfl_teams = [i.lower() for i in nfl_teams]
```

```python
for i in range(len(nfl_teams)):
    nfl_teams[i] = "-".join(nfl_teams[i].split(" "))
nfl_teams
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
nfl_df = pd.DataFrame()
for team_name in nfl_teams:
    url = "https://www.spotrac.com/nfl/"+ team_name + "/cap/2021/"
    page = requests.get(url)
    #soup.find('title')
    soup = BeautifulSoup(page.content, 'html.parser')
    # Look at div/table that has all active players info
    results = soup.find("div",{"class":'rtable-wrapper'})
    # Create player positions list by scraping data
    pos = []
    for i in results.find_all('span', class_='cap'):
        try:
            if i['title'] == '' and i.text != '0-':
                pos.append(i.text)
        except:
            continue
    pos = pos[:-1]

    # Get salary hit list by scraping data
    capHit = []
    dupCount = 0
    for i in results.findAll('span'):
        try:
            if i['title'] == 'Cap Hit':
                if dupCount %2 == 0:
                    capHit.append(i.text[1:])
                dupCount += 1
        except:
            continue

    AllCapHit = []
    AllCapHit.append(capHit[-1])
    capHit = capHit[:-1]

    # Get player name list by scraping data
    sCount = 0
    players = []
    for i in results.find_all('a'):
        if not i.text.endswith(')'):
            players.append(i.text)

    len(players)

    # Create new dataframe by merging all lists
    if nfl_df.empty:
        nfl_df = pd.DataFrame(zip(players,pos,capHit),columns=['player_name','pos','cap_hit'])
        
        nfl_df['team_name'] = team_name
        print(nfl_df)
    else:
        team_df = pd.DataFrame(zip(players,pos,capHit),columns=['player_name','pos','cap_hit'])
        team_df['team_name'] = team_name
        nfl_df = pd.concat([nfl_df,team_df],ignore_index=True)

#     nfl_df['team_name'] = 'Arizona Cardinals'
#     nfl_df
```

```python
nfl_df.team_name = nfl_df.team_name.str.split("-").str.join(" ").str.title()
```

```python
nfl_df.cap_hit = nfl_df.cap_hit.str.split(',').str.join("")
```
